```python
df_filled = df.fillna(0)
```

```python
### fill missing with mean of column
df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean())
df
```

```python
##Renaming column
df.rename(columns={'Date': 'Sales Date'}, inplace=True)
df
```

```python
#change datatype
df['Value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df['New Value'] = df['Value'].apply(lambda x: x*2)
df.head()
```

```python
## Data aggregateing and grouping
df.head()
```

```python
grouped_mean = df.groupby('Product')['Value'].mean()
grouped_mean
```

```python
df.groupby(['Product', 'Region'])['Sales'].sum()
```

```python
sum_value_by_product_region = df.groupby(['Product', 'Region'])['Sales'].sum().reset_index()
sum_value_by_product_region
```

```python
grouped_agg = df.groupby('Region')[['Value', 'Sales']].agg(['mean', 'sum', 'min', 'max','count'])
grouped_agg
```

```python
#Merging and joing dataframe
df1 = pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','D'],'Value2':[4,5,6]})

df1
```

```python
df2
```

```python
#MERGE ON KEY COLUMNS
pd.merge(df1, df2, on='Key',how='inner')
```

```python
pd.merge(df1, df2, on='Key',how='outer')
```

```python
pd.merge(df1, df2, on='Key',how='left')
```

```python
pd.merge(df1, df2, on='Key',how='right')
```
